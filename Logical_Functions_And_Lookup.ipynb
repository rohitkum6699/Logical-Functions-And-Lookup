#Q1) Write a formula to check if the marks in cell B2 are greater than or equal to 40. If yes, return "Pass"otherwise return"Fail".

| Student Name | Marks | Result |
|--------------|-------|--------|
| Riya         | 45    | Pass   |
| Siya         | 72    | Pass   |
| Raju         | 55    | Pass   |
| Rahul        | 80    | Pass   |
| Ram          | 30    | Fail   |


# Ans-->
```excel
=IF(B2>=40,"Pass","Fail")
```

### Explanation:
- `B2` → the cell containing the marks (adjust based on your sheet).
- `>=40` → condition to check if marks are greater than or equal to 40.
- `"Pass"` → returned if the condition is true.
- `"Fail"` → returned if the condition is false.




#Q2) A company gives a bonus  if sales are above 40,000 and attendance is greater than or equal to 90%. Write a formula to calculate whether the employee will get a bonus or not .
| Employee | Sales | Attendance | Result   |
|----------|-------|------------|----------|
| Arjun    | 55000 | 95         | Bonus    |
| Simran   | 65000 | 85         | No Bonus |
| Rohit    | 30000 | 90         | No Bonus |
| Anjali   | 25000 | 65         | No Bonus |

#Ans-->

### Formula:
```excel
=IF(AND(B2>40000,C2>=90),"Bonus","No Bonus")
```

### Explanation:
- `B2` → Sales column  
- `C2` → Attendance column  
- `AND(B2>40000,C2>=90)` → checks both conditions together  
- `"Bonus"` → returned if both conditions are true  
- `"No Bonus"` → returned if either condition fails  



# Q3) Write a nested IF formula:If marks in  B2 are ≥ 90 →  "A" , ≥75 → "B"  , ≥ 50 → "C" , otherwise "fail"

#Ans-->
### Grading Rules
- Marks ≥ 90 → `"A"`
- Marks ≥ 75 → `"B"`
- Marks ≥ 50 → `"C"`
- Otherwise → `"Fail"`

| Student | Marks | Result |
|---------|-------|--------|
| Vivek   | 85    | B      |
| Neha    | 96    | A      |
| Raj     | 57    | C      |
| Rani    | 92    | A      |
| Sneha   | 75    | B      |


### Formula
```excel
=IF(B2>=90,"A",IF(B2>=75,"B",IF(B2>=50,"C","Fail")))
```

### Explanation
- `IF(B2>=90,"A",...)` → If marks are 90 or above, return **A**.
- If not, Excel checks the next condition: `IF(B2>=75,"B",...)`.
- If not, Excel checks `IF(B2>=50,"C","Fail")`.
- If none of the conditions are met, it returns **Fail**.



# Q4) Find the Department of Employee ID = E103. Also explain which LOOKUP function is used and why?

# Ans-->



| EmpID | Name   | Department |
|-------|--------|------------|
| E101  | Raju   | IT         |
| E102  | Riya   | HR         |
| E103  | Roshni | Finance    |
| E104  | Sejal  | Sales      |

---

##  Step 1: Find Department of Employee ID = **E103**
Looking at the table:
- **E103 → Roshni → Finance**

So, the department is **Finance** .

---

##  Step 2: Which LOOKUP Function to Use?
In Excel, there are multiple lookup functions. The best choice here is **VLOOKUP**.

###  Why VLOOKUP?
- **VLOOKUP** searches for a value in the **first column** of a table (here, EmpID).
- It then returns a value from another column in the same row (here, Department).
- It’s simple and direct when your lookup column (EmpID) is on the left.

---

##  Step 3: Formula Example
To find the department of **E103**, you’d use:

```excel
=VLOOKUP("E103", A2:C5, 3, FALSE)
```

### Explanation:
- `"E103"` → the Employee ID you’re searching for  
- `A2:C5` → the table range (EmpID, Name, Department)  
- `3` → the column number of Department (third column in the range)  
- `FALSE` → ensures **exact match**  

Result → **Finance**



# Q5) Company has set the following commission structure for its sales team:

##  Commission Structure

| Sales Amount (≥) | Commission Rate |
|------------------|-----------------|
| 0                | 5%              |
| 20,000           | 10%             |
| 50,000           | 15%             |
| 80,000           | 20%             |

#If a salesperson achieves sales of ₹45,000, use the VLOOKUP function to determine the commission rate applicable.
#Ans-->
## Step 1: Sales Achieved
Salesperson’s sales = **₹45,000**

---

##  Step 2: Apply VLOOKUP
We’ll use **VLOOKUP with approximate match** because this is a **slab-based structure**.

Formula:

```excel
=VLOOKUP(45000, A2:B5, 2, TRUE)
```

### Explanation:
- `45000` → Sales amount to check  
- `A2:B5` → Commission slab table (Sales Amount & Rate)  
- `2` → Return the 2nd column (Commission Rate)  
- `TRUE` → Finds the **closest lower match** (needed for slabs)  

---

## Step 3: Result
- ₹45,000 falls between **20,000 and 50,000**  
- Closest lower match = **20,000 → 10%**  

Commission Rate applicable = **10%**

---

## Step 4: Commission Value (Optional)
If you want the actual commission amount:

```excel
=45000 * VLOOKUP(45000, A2:B5, 2, TRUE)
```

= ₹45,000 × 10% = **₹4,500**

---

 **Final Answer:**  
For sales of **₹45,000**, the commission rate is **10%**, calculated using **VLOOKUP with approximate match**.

---

## Step 3: Explanation
- `MATCH(D2, A2:A5, 0)` → Finds the position of **E105** in the EmpID column.  
  - E105 is the 2nd item → returns **2**.  
- `INDEX(B2:B5, 2)` → Returns the 2nd value from Salary column → **72,000**.  

---

##  Final Answer
The salary of **EmpID = E105** is **₹72,000**, found using the **INDEX-MATCH** function.


#Q8) Use the XLOOKUP function to find the Department of the employee with Emp ID = E303.
| EmpID | Department |
|-------|------------|
| E101  | HR         |
| E303  | IT         |
| E106  | Sales      |
| E301  | Finance    |


#Ans-->
##  Step 1: Goal  
We want the **Department of EmpID = E303**.

---

## Step 2: XLOOKUP Formula  
Suppose:
- EmpID values are in column **A** (A2:A5)  
- Department values are in column **B** (B2:B5)  
- Lookup EmpID (E303) is in cell **D2**  

Formula:

```excel
=XLOOKUP(D2, A2:A5, B2:B5, "Not Found")
```

---

##  Step 3: Explanation  
- `D2` → Lookup value (E303)  
- `A2:A5` → Lookup array (EmpID column)  
- `B2:B5` → Return array (Department column)  
- `"Not Found"` → Optional message if EmpID doesn’t exist  

---

##  Result  
For **EmpID = E303**, the department is **IT**.
